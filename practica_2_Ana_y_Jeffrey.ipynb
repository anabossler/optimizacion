Problema: Un fabricante de bebidas produce dos productos, A y B. El proceso de producción requiere dos recursos, M y N. El fabricante desea maximizar sus beneficios, pero también debe tener en cuenta las limitaciones de producción.

Datos del problema:

Beneficio por unidad de A: 100
Beneficio por unidad de B: 150
Recurso M disponible: 200 unidades
Recurso N disponible: 300 unidades
Requerimientos de M por unidad de A: 1
Requerimientos de M por unidad de B: 2
Requerimientos de N por unidad de A: 2
Requerimientos de N por unidad de B: 1
Formulación del problema binivel:

Primer nivel: Maximizar Z = 100A + 150B (beneficio total)
sujeto a:

A + 2B ≤ 200 (restricción de recurso M)
2A + B ≤ 300 (restricción de recurso N)
Segundo nivel (Dualidad Fuerte): Minimizar W = 200y1 + 300y2 (costo total)
sujeto a:

y1 + 2y2 ≥ 100
2y1 + y2 ≥ 150

```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp

# Resolver el problema del primer nivel (con restricciones incluidas)
A = cp.Variable(pos=True)
B = cp.Variable(pos=True)
# Creamos dos variables llamadas A y B para representar la cantidad de productos A y B que queremos producir. 
# Estas variables son siempre positivas (o cero).



Z = 100 * A + 150 * B 
# Calculamos el beneficio total (Z) multiplicando la cantidad de productos A y B por sus respectivos beneficios (100 y 150).

constraints = [
    A + 2 * B <= 200,
    2 * A + B <= 300,
]

# Establecemos las restricciones del problema. No podemos usar más de 200 unidades del recurso M y 300 unidades del recurso N.
#################################
problem1 = cp.Problem(cp.Maximize(Z), constraints)
problem1.solve()

# Creamos el primer problema de optimización, donde queremos maximizar el beneficio total (Z) cumpliendo las restricciones.
# Resolvemos el primer problema de optimización para encontrar la mejor cantidad de productos A y B a producir.

####################################

# Preparar datos para graficar
a = np.linspace(0, 200, 1000)
b1 = (200 - a) / 2
b2 = 300 - 2 * a

plt.plot(a, b1, label="A + 2B <= 200")
plt.plot(a, b2, label="2A + B <= 300")
plt.fill_between(a, np.minimum(b1, b2), 0, where=(b1 >= 0) & (b2 >= 0), alpha=0.3)

# Graficar la solución óptima
plt.scatter(A.value, B.value, color="red", label="Solución óptima")

plt.xlim(0, 200)
plt.ylim(0, 200)
plt.xlabel("Producto A")
plt.ylabel("Producto B")
plt.legend()
plt.title("Optimización y solución gráfica")
plt.show()
```

Primer nivel (Problema primal):

Maximizar:
$$
Z = 100A + 150B
$$

Sujeto a:

$$
A + 2B \le 200 \\
2A + B \le 300 \\
A \ge 0 \\
B \ge 0
$$


Minimizar:

$$
W = 200y_1 + 300y_2
$$

 
Sujeto a:
$$
y_1 + 2y_2 \ge 100 \\
2y_1 + y_2 \ge 150 \\
y_1 \ge 0 \\
y_2 \ge 0
$$

Conclusiones:

-- El problema planteado es un problema de programación lineal binivel que busca maximizar el beneficio total de producir dos productos (A y B) mientras se tienen en cuenta las restricciones de recursos (M y N).

-- Las restricciones de recursos establecen límites en la cantidad de productos A y B que se pueden producir, basados en la disponibilidad de recursos M y N.

-- La solución óptima del problema primal nos proporciona la combinación de productos A y B que maximizan el beneficio total, cumpliendo con las restricciones de recursos.

-- La solución óptima del problema dual nos proporciona información sobre el costo total asociado con los recursos M y N para alcanzar el beneficio máximo.

-- La dualidad fuerte en este caso nos permite encontrar la solución óptima del problema primal a través de la solución óptima del problema dual. En otras palabras, si encontramos la mejor manera de minimizar el costo total de los recursos, también encontraremos la mejor manera de maximizar el beneficio total.
